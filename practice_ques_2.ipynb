```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import sum, collect_list, struct

# spark = SparkSession.builder.appName("PersonItemWeight").getOrCreate()

data = [
    ("alice", "carrot", 1),
    ("bob", "banana", 3),
    ("alice", "tomato", 4),
    ("bob", "carrot", 2),
    ("charlie", "banana", 5),
    ("bob", "apple", 1),
    ("alice", "tomato", 2),
    ("charlie", "carrot", 3)
]

cols = ["name", "item", "weight"]

df = spark.createDataFrame(data, cols)
```

```python
df.show()
```

```python
from pyspark.sql.functions import *
aggregated_df = df.groupBy(col("name"), col("item")).agg(sum(col("weight")).alias("total_weight"))
```

```python
aggregated_df.show()
```

```python
final_result = aggregated_df.groupBy(col("name")).agg(collect_list(struct("item" , "total_weight")).alias("total_posession"))
final_result.show(20, False)
```
